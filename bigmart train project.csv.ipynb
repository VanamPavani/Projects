```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection  import cross_val_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
feat = ['Outlet_Size','Outlet_Type','Outlet_Location_Type','Item_Fat_Content',"Item_Type"]
X = pd.get_dummies(bigmart_train[feat])
bigmart_train = pd.concat([bigmart_train,X],axis=1)
```

```python
bigmart_train.head()
```

```python
bigmart_test=pd.read_csv("bidatamart.testfile.csv")
test1 = bigmart_test.copy()
```

```python
bigmart_test.drop(labels = ["Outlet_Establishment_Year"],inplace = True,axis =1)
```

```python
feat = ['Outlet_Size','Outlet_Type','Outlet_Location_Type','Item_Fat_Content',"Item_Type"]
X1 = pd.get_dummies(bigmart_test[feat])
bigmart_test = pd.concat([bigmart_test,X1],axis=1)
```

```python
map1 = {"Small":1,"Medium":2,"High":3}
bigmart_test["Outlet_Size"] = bigmart_test["Outlet_Size"].map(map1)
bigmart_test["Item_Weight"] = bigmart_test["Item_Weight"].fillna(bigmart_test.Item_Weight.mean())
bigmart_test["Outlet_Size"] = bigmart_test["Outlet_Size"].fillna(bigmart_test["Outlet_Size"].median())
```

```python
bigmart_train.drop(labels = ["Outlet_Size",'Outlet_Location_Type',"Outlet_Type",'Item_Fat_Content','Outlet_Identifier','Item_Identifier',"Item_Type"],axis=1,inplace = True)
bigmart_test.drop(labels = ["Outlet_Size",'Outlet_Location_Type',"Outlet_Type",'Item_Fat_Content','Outlet_Identifier','Item_Identifier',"Item_Type"],axis=1,inplace = True)
```

```python
X_train = bigmart_train.drop(labels = ["Item_Outlet_Sales"],axis=1)
y_train = bigmart_train["Item_Outlet_Sales"]
X_train.shape,y_train.shape
```

```python
bigmart_train.head()
```

```python
from sklearn import preprocessing

x = X_train.values #returns a numpy array
test_s = bigmart_test.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled_train = min_max_scaler.fit_transform(x)
x_scaled_test = min_max_scaler.fit_transform(test_s)
df_train = pd.DataFrame(x_scaled_train)
df_test = pd.DataFrame(x_scaled_test)
```

```python
df_train.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train, y_train, test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
plt.scatter(y_test, preds)
plt.show()
```

```python
sns.distplot((y_test-preds),bins=50)
plt.show()
```

```python
from sklearn import metrics
print("MAE:", metrics.mean_absolute_error(y_test, preds))
print('MSE:', metrics.mean_squared_error(y_test, preds))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, preds)))
```
